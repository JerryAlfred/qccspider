```python

df = pd.read_excel("2020.xlsx")

df
```

```python
qieci_lst = []

# for i in (df.反馈情况):
#     if i == i:
#         qieci_lst.append(jieba.lcut(i))

for i in (df.行业备注):
    if i == i:
        qieci_lst.append(jieba.lcut(i))
        
words = [j for i in qieci_lst for j in i]

counts = {}

for word in words:
    if len(word) == 1:
        continue
    else:
        counts[word] = counts.get(word, 0) + 1 

items = list(counts.items())    

items.sort(key=lambda x: x[1], reverse=True)      

for i in range(len(items)):
    word, count = items[i]
#     print("{0:<5}->{1:>5}".format(word, count))

freq = pd.DataFrame()

freq['word'] = word
freq['count'] = count

for i in range(len(items)):
    
    freq.loc[i] = [items[i][0], items[i][1]]
```

```python
freq.to_excel('2020项目切词的频率_行业备注.xlsx')
```
